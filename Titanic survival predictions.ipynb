```python
df1=df.drop(columns=['Name','Cabin','Fare','Ticket','SibSp','Parch'])
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.replace({'Sex': {'female': 0, 'male': 1}, 'Embarked': {'S': 0, 'C': 1, 'Q': 2}}, inplace=True)
```

```python
df1
```

```python
from sklearn.impute import KNNImputer
# Initialize KNNImputer
imputer = KNNImputer(n_neighbors=2)

# Fit and transform the data
imputed_data = imputer.fit_transform(df1[['Age', 'Sex', 'Pclass']])

# Create a DataFrame from the imputed data
df_imputed = pd.DataFrame(imputed_data, columns=['Age', 'Sex', 'Pclass'])

# Update the original DataFrame with the imputed 'Age' column
df1['Age'] = df_imputed['Age']

print(df1)
```

```python
df1
```

```python
df1.info()
```

```python
df1=df1.dropna()
```

```python
df1.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
x=df1.iloc[:,[2,3,4,5]]
y=df1.iloc[:,1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#classifier=KNeighborsClassifier(n_neighbors=7,algorithm='kd_tree',weights='uniform')
#classifier=LogisticRegression()
#classifier=RandomForestClassifier()
models={
    'Logistic Regression':LogisticRegression(),
    'KNNeighbors Classifier':KNeighborsClassifier(),
    'Random Foresr Classifier':RandomForestClassifier(),
    'Gaussian Naive Bayes': GaussianNB(),
    'Multinomial Naive Bayes':MultinomialNB(),
    'Bernoulli Naive Bayes':BernoulliNB(),
    'Decision Tree Classifier': DecisionTreeClassifier()
}
```

```python
r=[]
for name,model in models.items():
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    ac=accuracy_score(y_pred,y_test)
    r.append({
    'Name of the model':name,
    'Accuracy':ac
    })
```

```python
df2=pd.DataFrame(r)
```
